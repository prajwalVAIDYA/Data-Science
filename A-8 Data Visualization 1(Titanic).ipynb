# DATA VISUALIZATION - I

1. Use the inbuilt dataset 'titanic'. The dataset contains 891 rows and contains information about the passengers who boarded the unfortunate Titanic ship. Use the Seaborn library to see if we can find any patterns in the data.
2. Write a code to check how the price of the ticket (column name: 'fare') for each passenger is distributed by plotting a histogram.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df
```

```python
import warnings
warnings.filterwarnings('ignore')
```

# Distribution Plots


## 1. distplot

```python
sns.distplot(x=df['age'],bins=10)
```

```python
sns.distplot(x=df['age'],bins=10,kde=False)
```

## 2. Joint Plot

```python
sns.jointplot(x=df['age'],y=df['fare'],kind='scatter')
```

```python
sns.jointplot(x=df['age'],y=df['fare'],kind='hex')
```

## 3. Rug Plot

```python
sns.rugplot(df['fare'])
```

# Categorcal Plot


## 1. Bar Plot

```python
sns.barplot(x='sex',y='age',data=df)
```

```python
sns.barplot(x='sex',y='age',data=df,estimator=np.std)
```

## 2.Count Plot

```python
sns.countplot(x='sex',data=df)
```

## 3. Box Plot

```python
sns.boxplot(x=df.sex,y=df.age)
```

```python
sns.boxplot(x=df.sex,y=df.age,hue=df.survived)
```

## 4. Violin Plot

```python
sns.violinplot(x=df.sex,y=df.age)
```

```python
sns.violinplot(x=df.sex,y=df.age,hue=df.survived)
```

# Advanced Plot


## 1.Strip Plot

```python
sns.stripplot(x=df.sex,y=df.age,jitter=False)
```

```python
sns.stripplot(x=df.sex,y=df.age,jitter=True)
```

```python
sns.stripplot(x=df.sex,y=df.age,jitter=True,hue=df.survived)
```

## 2. Swarm Plot

```python
sns.swarmplot(x=df.sex,y=df.age)
```

```python
sns.swarmplot(x=df.sex,y=df.age,hue=df.survived)
```
