Use the inbuilt dataset 'titanic' as used in the above problem. Plot a box plot for distribution
of age with respect to each gender along with the information about whether they survived or

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset('titanic')
df
```

## Box Plot

```python
sns.boxplot(x='sex', y='age', data=df)
```

```python
sns.boxplot(x=df.sex,y=df.age,hue=df.survived)
```

```python
def pieChart(df,categoricalColumn):
  plotNo = 1
  if df[categoricalColumn].dtype == 'O':
    plt.subplot(1, plotNo, plotNo)
    df[categoricalColumn].value_counts().plot(kind='pie', autopct='%1.1f%%')
    plotNo+=1
    plt.plot()
```

```python
df['sex'].value_counts().plot(kind='pie', autopct='%1.1f%%')
```
